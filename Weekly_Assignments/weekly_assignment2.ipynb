***Weekly Assignment: Quiz Game***

**Problem Statement:**
Create a simple command-line quiz game that reads questions from a file, allows users to answer them, and provides a score at the end. The questions and answers should be stored in a text file.

**Question Class:**

Create a Question class with attributes like question_text and correct_answer. Include methods to initialize the question and check if a provided answer is correct.

**File Handling:**

Use a text file (quiz_questions.txt) to store questions and answers. Each line in the file represents a question with the format: <question_text>|<correct_answer>.

**Read Questions:**

Implement a function to read questions from the file and create instances of the Question class for each question.

**Quiz Game Loop:**

Create a loop that presents each question to the user, allows them to input their answer, checks if it's correct, and keeps track of their score.

**Score Calculation:**

Calculate and display the user's score at the end of the quiz.

**Error Handling:**

Implement error handling to handle cases where the file is not found or if there is an issue reading its content.

**Negative Marking:**

If there are 3 wrong answers in the quiz, 1 mark will be deducted.
There will be no negative marking for one-word answers.

```python
from IPython.display import clear_output
import time


class InvalidOptionError(Exception):                                            # Custom Exception
    "raised when invalid inputs are given"
    pass


class NoAnswerEnteredError(Exception):                                           # Custom Exception
    "raised when no answer is given"
    pass


class question():
    score = 0
    negative_score = 0
    dict = {}
    isMcq = True
    Total_correct = 0
    Total_wrong = 0    

    # this is the init function it do nothing
    def __init__(self):     
        pass  
    

    @classmethod
    def wrong_right(cls,correct,wrong):
        cls.Total_correct += correct
        cls.Total_wrong += wrong

    def display_question(self,file1):                                      #this fuction is used to display questions only
        i = 1                                                                   # this is used to run readline function for one time
        
        while(i):   
            data = file1.readline()                                             # this is for to clean the data
            # data = file1.readline().replace("\n","")                          # this is for to clean the data    
            data = data.replace('\r','')                                        # this is for to clean the data
            # data = data.lstrip('')                                            # this is for to clean the data
            data = data.replace('\n','')                                        # this is for to clean the data   
            
            if(data!=""):
                # print(list(data.lower()))
                if(data.lower() == "one word answer"):                            # checking is the one marks question started for not
                    print("Now One Word Question")
                    question.change_type()
                else:                                                       # this is used to check it the data that come after cleaning is not empty
                    print(data)
                    i-=1
                    print()
        if(question.isMcq==True):                                                          # used to check is the question is MCQ or
            self.display_options(file1)                                                 # calling the function that will display the options
        else:
            self.assign_marks_for_oneword(file1)                                       # if it is mcq it will call display option functions
        
    @classmethod
    def change_type(cls):                                                       # class method to change the question type to not mcq so that options fucntion not run 
        cls.isMcq = False

    
    def display_options(self,file1):                                            # fuction for showing questions
        i = 4                                                                   # there are 4 option it will read 4 lines
        char  = 'a'                                                             # option starting from a then b then c and so on upto d
        self.dict = {}
        while(i):
            data = file1.readline()                                             #cleaning data
            data = data.replace('\r','')                                        #cleaning data
            data = data.strip(' ')                                              #cleaning data
            data = data.replace('\n','')                                        #cleaning data 
            data = data.rstrip(" ")                                             #cleaning data
     
            
            if(data!=""):                                                       # this is used to check it the data that come after cleaning is not empty
                self.dict[char] = data                                        # inserting data into dictionary
                print(f"({char}) {''.join(data)}")
                char = chr(ord(char) + 1)                                       #changing option character    
                i-=1
                print()
        self.assign_marks_for_mcq(file1)                                        # calling the assign marks fucntion for mcq type questions
     


    def enter_solution_for_mcq(self):                                            #this function will take input and return the option that user will enter
        while(True):
            try:                                                
                self.entered_option = input("Enter the mcq option")
                if(self.entered_option!='a' and self.entered_option!='b' and self.entered_option!='c' and self.entered_option!='d' ):
                    raise InvalidOptionError("Invalid Options")
            except InvalidOptionError as Arguments:
                print(Arguments)
                print("Please Enter Valid Option")
            else:

                return self.entered_option
        
    def assign_marks_for_mcq(self,file1):                                                # used to check whether the user enter the right answer and according assign the marks
        i = 1                                                                              #used to print only one line
        option_selected = self.enter_solution_for_mcq()                                 # its for storing the entered options 
        while(i):       
            data = file1.readline()                                                         
            data = data.replace('\r','')                                                #cleaning the data
            final_data = data.replace('\n','')                                          #cleaning the data
            final_data = final_data.lstrip('')                                          #cleaning the data

            if(final_data!=""):
                i-=1
                final_data = final_data.split(':')[1]
                final_data = final_data.lstrip(' ')
                if(self.dict[option_selected].lower() == final_data.lower()):
                    print("Great!!! That's right")
                    question.wrong_right(1,0)                                                  #Positive marks
                    self.score += 1             
                else:
                    print("Opps!!! That's Wrong")
                    question.wrong_right(0,1)
                    self.negative_score -= 0.25                                               # negative marks

            question.calculateScore(self)
                

    def enter_solution_for_oneword(self):                                                   #it's used for entering the one word answer
        while(True):
            try:                                                
                self.answer_entered = input("Enter the your options")                        # Exceptional handling for handling the case like empty string
                if(len(self.answer_entered)<1):     
                    raise NoAnswerEnteredError("Enter atleast one Character")
            except NoAnswerEnteredError as Arguments:
                print(Arguments)
             
            else:

                return self.answer_entered


    def assign_marks_for_oneword(self,file1):                                                # used to check whether the user enter the right answer and according assign the marks
        i = 1                                                                               # used to print only one line
        answer_entered = self.enter_solution_for_oneword()                                 # its for storing the entered options 
        while(i):                                                                                   
            data = file1.readline()                                                         
            data = data.replace('\r','')                                                    #cleaning the data                                                          
            final_data = data.replace('\n','')                                              #cleaning the data      
            final_data = final_data.lstrip('')                                              #cleaning the data      

            if(final_data!=""):                                                             #checking if the line is not empty
                i-=1
                # final_data = final_data.split(':')[1]
                final_data = final_data.replace(" ","").lower()
                # print(answer_entered)
                # print(final_data)
                if(answer_entered.replace(" ","") == final_data):                           #checking is the entered value is equal to correct value

                    print("Very Good!!! That's right")
                    question.wrong_right(1,0)                                               #Assigning the marks
                    self.score += 2 
                else:
                    print("Opps!!! That's Wrong")                               
                    question.wrong_right(0,1)                                               #Assigning the marks
            
            question.calculateScore(self)

       
                
    @classmethod                                                                #used to calculate the score and store it in class variable
    def calculateScore(cls,self):
        cls.score = self.score + self.negative_score

    @classmethod
    def display_result(cls,n):                                                  #displaying the result of the quiz
        print(f"""
                    -----------
                    Result Page
                    ------------
              
            Total Number of Questions => {n}
            Total Correct Questions => {question.Total_correct}
            Total Wrong Questions => {question.Total_wrong}

                -----------------------
                | You Scored => {question.score}/35 |
                -----------------------

              KEEP PUSHING YOURSELF !!! 💪

                """)


with open("extra_files/mcq_quiz.txt",'r',newline="") as file1:
    n = 30                                                                                  # No. of Questions
    print("Your test is going to start in 3 sec")
  
    while(n):
    
        time.sleep(1)                                                                       #used for 1 sec delay between questions
        clear_output(wait=True)                                                             # used to clear console output so that it look clean
        obj1 = question()
        obj1.display_question(file1)
        n-=1
        
        
    clear_output(wait=True)
    question.display_result(n)                                                              # Displaying the result
        
```
