 1: Read Total profit of all months and show it using a
line plot
Total profit data provided for each month. Generated line plot must include the following
properties:
● X label name = Month Number
● Y label name = Total profit 

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("extra_files/company_sales_data.csv")
# print(df)
x = df["month_number"]
y = df["total_profit"]

plt.plot(x,y)
plt.title("Company Profit Per Month")
plt.xlabel("Month Number")
plt.ylabel("Profit in Dollar")
plt.xticks(x)
plt.show()
```

2: Get total profit of all months and show line plot with
the following Style properties

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("extra_files/company_sales_data.csv")
# print(df)
x = df["month_number"]
y = df["total_profit"]

plt.plot(x,y,"--",marker='.',color='r',markerfacecolor = "black",markersize = 13,linewidth = 3,label="Profit data of last year")
plt.title("Company Profit Per Month")
plt.xlabel("Month Number")
plt.ylabel("Profit in Dollar")
plt.xticks(x)
plt.legend(loc="lower right")
plt.show()
```

8: Calculate total sale data for last year for each
product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.
The Pie chart should look like this:

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("extra_files/company_sales_data.csv")

label = ["Toothpaste","facewash","facecream","Bathing Soap","Shampoo","Moisturizer"]
data = [df["toothpaste"].sum(),df["facewash"].sum(),df["facecream"].sum(),df["bathingsoap"].sum(),df["shampoo"].sum(),df["moisturizer"].sum()]
plt.pie(data, labels = label,autopct='%1.1f%%')

plt.title("Sales data")
plt.legend(loc="lower right")
plt.xticks()
plt.show()
```
