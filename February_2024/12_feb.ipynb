```python
import matplotlib.pyplot as plt 
import numpy as np



x = ["Python","C","C++","Java"]
y = [85,70,60,82]
z = [20,30,40,50]
i = np.arange(len(x))

width = 0.2
plt.bar(i,y,width,color = 'r',align="center",label = "Popularity")
plt.bar(i+width,z,width,color = 'g',label = "Popularity1")

plt.xlabel("Languages",fontsize=15)
plt.ylabel("Popularity",fontsize=15)
plt.title("Dipesh's Chart",fontsize=15)
plt.xticks(i+width/2,x)


plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = [0,1,2,3,4]
# y = [0,2,4,6,8]
z = [x*2 for x in x]
print(z)
plt.plot(x,z,marker="^",linestyle="dashed",label="2x")
z = [i**2 for i in x]
plt.plot(x,z,color= 'r',label="X^2")

plt.xticks(x)

plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend()
plt.title("Our First Graph")
plt.show()
```
